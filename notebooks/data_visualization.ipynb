```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Example DataFrame creation:
# df = pd.DataFrame({'orig_lang': ['English', 'Spanish', 'French', 'German', 'English', 'French', 'Spanish']})

# Plotting a simple bar chart with sorted values
df['orig_lang'].value_counts().sort_values().plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Original Language')
plt.ylabel('Count')
plt.title('Distribution of Original Languages')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df['orig_lang'] contains the languages
labels = df['orig_lang'].value_counts().index
sizes = df['orig_lang'].value_counts().values

plt.figure(figsize=(7, 3))  # Adjust the dimensions as needed
plt.pie(sizes, labels=None, autopct='', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.title('Distribution of Movies Accordting to the Languages')

# Add labels and percentages in the legend on the right-hand side
legend_labels = [f'{label} ({size})' for label, size in zip(labels, sizes)]
plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

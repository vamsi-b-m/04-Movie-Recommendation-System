```python
def recommend(movies, movie_count_to_recommend):
    index = data[data['names']==movies].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distances[1:movie_count_to_recommend]:
        print(data.iloc[i[0]].names)




def recommend_with_genre(target_genre):
    start_column_index = 3
    movie_names = []
    for movie in range(len(data)):
        movie_row = data.iloc[movie, start_column_index:]
        if movie_row[target_genre] and not movie_row[movie_row.index != target_genre].any():
            movie_names.append(data.iloc[movie]['names'])


    movie_count = len(movie_names)
    movie_count_to_recommend = 15 - movie_count
    if movie_count <= 10:
        {print(movie) for movie in movie_names}
        recommend(movie_names[0], movie_count_to_recommend)
    else:
        {print(movie) for movie in movie_names[:15]}
```

```python
import requests, urllib

def get_movie_poster_url(movie_name):
    api_key = 'b1f9d8f7'  # Replace 'your_omdb_api_key' with your actual OMDB API key
    base_url = 'http://www.omdbapi.com/'
    params = {
        'apikey': api_key,
        't': movie_name
    }

    response = requests.get(base_url, params=params)
    data = response.json()

    if 'Error' in data:
        print(f"Error: {data['Error']}")
        return None

    return data.get('Poster', 'N/A')


def download_image(url, save_path):
    try:
        # Send a GET request to the URL to download the image
        response = requests.get(url)
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Open a file and write the content of the response into it
            with open(save_path, 'wb') as file:
                file.write(response.content)
            print(f"Image downloaded successfully and saved as {save_path}")
        else:
            print(f"Failed to download image from {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    movie_name = "Maya Bazar"
    poster_url = get_movie_poster_url(movie_name)
    download_image(url=poster_url, save_path=movie_name+".jpg")
    
    if poster_url != 'N/A':
        print("Poster URL:", poster_url)
    else:
        print("Poster not found.")
```
